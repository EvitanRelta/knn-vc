```python
import torch, torchaudio

knn_vc = torch.hub.load('bshall/knn-vc', 'knn_vc', prematched=True, trust_repo=True, pretrained=True, device='cuda')
```

```python
import torch
import torchaudio
import torchaudio.transforms as T
import os

def convert_mp3_to_16khz_mono_wav(input_path: str, output_path: str):
    """
    Converts an audio file to a 16kHz, single-channel WAV file.

    Args:
        input_path: Path to the input audio file.
        output_path: Path to save the output WAV file.
    """
    # Load the MP3 file
    waveform, sample_rate = torchaudio.load(input_path)

    # Resample to 16kHz
    if sample_rate != 16000:
        resampler = T.Resample(orig_freq=sample_rate, new_freq=16000)
        waveform = resampler(waveform)
        sample_rate = 16000

    # Convert to mono
    if waveform.shape[0] > 1:
        waveform = torch.mean(waveform, dim=0, keepdim=True)

    # Save as WAV
    torchaudio.save(output_path, waveform, sample_rate)

    print(f"Converted {input_path} to {output_path}")

src_path = "data/good_morning.m4a"
src_wav_path = os.path.splitext(src_path)[0] + ".wav"
convert_mp3_to_16khz_mono_wav(src_path, src_wav_path)
```

```python
# path to 16kHz, single-channel, source waveform
# src_wav_path = 'data/5105.wav'
# list of paths to all reference waveforms (each must be 16kHz, single-channel) from the target speaker
ref_wav_paths = ['data/4507.wav', ]

query_seq = knn_vc.get_features(src_wav_path)
matching_set = knn_vc.get_matching_set(ref_wav_paths)
```

```python
out_wav = knn_vc.match(query_seq, matching_set, topk=4)
```
